```python
import pandas as pd
import math

data = {
"Gender": ['M','M','M','M','M','M','F','F','F','F','M','M','M','M','F','F','F','F','F','F'],
"Income": [40000,50000,52500,80000,100000,100000,42000,40000,50000,62000,80000,70000,65000,140000,100000,100000,120000,80000,82000,150000],
"Car Type": ['Family','Sports','Sports','Sports','Sports','Sports','Sports','Sports','Sports','Luxury','Family','Family','Family','Luxury','Luxury','Luxury','Luxury','Luxury','Luxury','Luxury'],
"Shirt Size": ['Small','Medium','Medium','Large','Extra Large','Extra Large','Small','Small','Medium','Large','Large','Extra Large','Medium','Extra Large','Small','Small','Medium','Medium','Medium','Large'],
"Class": ['C0','C0','C0','C0','C0','C0','C0','C0','C0','C0','C1','C1','C1','C1','C1','C1','C1','C1','C1','C1']
}

df = pd.DataFrame(data)
```

```python
def prior_prob(df, cls):
    return len(df[df['Class'] == cls]) / len(df)

def gaussian_pdf(x, mu, var):
    return (1 / math.sqrt(2 * math.pi * var)) * math.exp(-((x - mu)**2) / (2 * var))


def likelihood(df, feature, value, cls):
    rows = df[df['Class'] == cls]
    if(feature == "Income") :
        mean = rows["Income"].mean()
        variance = rows["Income"].var()
        return gaussian_pdf(value , mean , variance)
    return len(rows[rows[feature] == value]) / len(rows)
```

```python
def predict(features):
    # Priors
    p_c0 = prior_prob(df, 'C0')
    p_c1 = prior_prob(df, 'C1')
    
    # Likelihoods
    for col, val in features.items():
        p_c0 *= likelihood(df, col, val, 'C0')
        p_c1 *= likelihood(df, col, val, 'C1')
    
    # Compare final probabilities
    print(p_c0 , p_c1)
    return 'C0' if p_c0 > p_c1 else 'C1'

X1 = {"Gender": "M", "Income": 130000, "Car Type": "Luxury", "Shirt Size": "Medium"}
X2 = {"Gender": "F", "Income": 42000, "Car Type": "Sports", "Shirt Size": "Small"}

print("X1 Prediction:", predict(X1))
print("X2 Prediction:", predict(X2))
```
